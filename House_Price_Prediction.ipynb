```
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
import pandas as pd
import numpy as np

# To get data from drive is necessary
from google.colab import drive 
drive.mount('/content/drive')
#!ls '/content/drive/My Drive'
# Path of the file to read
test_data_file_path = "/content/drive/My Drive/Colab-data/test.csv"
home_data_file_path = "/content/drive/My Drive/Colab-data/train.csv"
home_data = pd.read_csv(home_data_file_path)
test_data = pd.read_csv(test_data_file_path)
features = ['LotFrontage', 'LotArea', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF']
X_test = test_data[features]
X_test.fillna(0, inplace = True)
```

```
# Create target object and call it y
y = home_data.SalePrice
# Create features object(X)
features = ['LotFrontage', 'LotArea', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF']
X = home_data[features]
X.fillna(0, inplace = True)
```

```
# Specify Model
home_data_model = DecisionTreeRegressor(random_state=1)
# Fit Model
home_data_model.fit(X, y)
```

```
# Make validation predictions and calculate mean absolute error
import numpy as np

val_predictions_appended = home_data_model.predict(X_test)
val_predictions = np.append(val_predictions_appended,[0])
val_mae = mean_absolute_error(val_predictions, y)
print("Validation MAE : {:,.0f}".format(val_mae))
```

```
# Mean Absolute Error function to find maximum leaf nodes
def get_mae(max_leaf_nodes, train_X, test_X, train_y, test_y):
  model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state = 0)
  model.fit(train_X, train_y)
  preds_val = model.predict(test_X)
  mae = mean_absolute_error(test_y, preds_val)
  return(mae)

candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500, 1000, 2000, 5000]
# Loop to find the ideal tree size
scores = {i : get_mae(i , X, X_test, y, val_predictions_appended) for i in candidate_max_leaf_nodes}
best_tree_size = min(scores, key = scores.get)
final_model = DecisionTreeRegressor(max_leaf_nodes = best_tree_size, random_state = 1)
final_model.fit(X, y)
```
